```python
import numpy as np
import matplotlib.pyplot as plt

##-----------------------------------Define the MDP components--------------------------------##

num_states = 10
num_actions = 5
states = [f's{i}' for i in range(num_states)]
actions = [f'a{i}' for i in range(num_actions)]

###------------------------ Transition probability matrix P[s, a, s']-------------------------###

# Random generation of matrix
def Markov_decision_matrix(num_states, num_actions):
  Markov = []
  for s in range(num_states):
    matrix = np.random.rand(num_actions,num_states)
    # Normalization of each row
    matrix = matrix/matrix.sum(axis=1, keepdims=True)
    Markov.append(matrix)
  return np.array(Markov)
np.random.seed(42)
P = Markov_decision_matrix(num_states, num_actions)




##--------------------------Reward function R[s, a]--------------------------------------##
R = np.round(np.random.rand(num_states, num_actions),2)  # random generation of R


# Action_value Function Q(initial value)
Q = np.zeros((num_states, num_actions))

# Discount factor
gamma = 0.7

Q.shape, R.shape, P.shape
```

```python
V = np.zeros(num_states)
#np.dot(P[a,:,:],V)
np.dot(P,V).shape # The product make really sense
```

```python
##------------ Definition of Bellman operator------------------------##

def Bellman_Operator(Q,R,P,gamma):
  V_Q = np.max(Q, axis=1)
  return R + gamma*np.dot(P,V_Q)

#Bellman_Operator(Q,R,P,gamma)# Test pour Q = Zeros_23
```

```python
##--------------------------Value iteration algorithm----------------------##

def Matrix_value_iteration(Q, P, R, gamma, theta=0.0001):
  iter = 0
  err = []
  while True:
    TQ=Bellman_Operator(Q,R,P,gamma)
    delta = np.max(np.abs(Q - TQ))
    err.append(delta)
    Q = TQ
    iter+=1
    if delta < theta:
      break
  V_star = np.max(Q, axis=1)
  return V_star, iter, err, R.shape
```

> *After convergence, the optimal policy is derived by selecting the action that maximizes the expected value for each state*.
$$\forall s\in \mathbb{S}, \pi^*(s) = \underset{a\in\mathbb{A}}{argmax}Q^*(s,a) = \underset{a\in\mathbb{A}}{argmax}\left[r(s,a)+\gamma\underset{s'}{\sum}P(s'/s,a)V_Q(s')\right].$$

```python
##----------------------Optimal policy---------------------------------##

def optimal_policy(P, R, V, gamma):
    num_states, num_actions, _ = P.shape
    policy = np.zeros(num_states, dtype=int)
    for s in range(num_states):
        policy[s] = np.argmax([
            R[s,a] + sum(gamma *P[s, a, s_next] * ( V[s_next])#np.argmax(Q[:,s])
                    for s_next in range(num_states))
                for a in range(num_actions)])
    return policy

policy_star = optimal_policy(P1, R1, optimal_values, gamma=0.9)
print("\nOptimal Policy:")
for i, action in enumerate(policy_star):
    print(f"π({states[i]}) = {actions[action]}")
```

```python
##-----------------------Plotting function---------------------##

def loss_Plot(k,err,gamma,sh,type_err):
    K = np.arange(k)
    if type_err == "log":
      plt.plot(K, np.log(err), label = f"$\gamma$={gamma}:({k} iterations)")
      plt.xlabel("Iterations")
      plt.ylabel("$\log_{(||Q^k - TQ^k||)}$")
      plt.title(f"S = {sh[0]}, A = {sh[1]}")
      plt.grid(visible=True)
      plt.legend()
    else:
      plt.plot(K, err, label = f"$\gamma$={gamma}:({k} iterations)")
      plt.xlabel("Iterations")
      plt.ylabel("Error")
      plt.title(f"S = {sh[0]}, A = {sh[1]}")
      plt.grid(visible=True)
      plt.legend()
    return 0
```

```python
import time
###----------------------Analyze of convergence based on gamma's value--------------------###
Gamma = [1-1e-1, 1-1e-2, 1-1e-3, 1-1e-4]
Times = []
for gam in Gamma:
  start = time.time()
  _, k, err,sh = Matrix_value_iteration(Q,P, R, gam)
  end = time.time()
  Times.append((end-start))
  loss_Plot(k,err,gam,sh,"log")
```

```python
###----------------------Plotting of Gamma/times curve--------------###
Gamma_log = [np.log10(1/(1-g)) for g in Gamma] # Log(times)
Times_log = [np.log10(t) for t in Times] # Log(1/(1-gamma))
plt.figure()
for i in range(len(Times)):
  plt.text(Gamma_log[i],Times_log[i],f'{Gamma[i],round(Times[i],5)}')

plt.plot(Gamma_log,Times_log, "-*", label="$(\gamma,t=f(\gamma))$")
plt.xlabel("$\log[\dfrac{1}{1-\gamma}]$")
plt.ylabel("$\log(t)$")
plt.title("Times to compute $V^*$ when $\gamma$ increase ")
plt.legend()
plt.grid(1)
```
