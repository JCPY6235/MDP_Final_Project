```python

##-----------------------------------Define the MDP components--------------------------------##

num_states = 20
num_actions = 10

###------------------------ Transition probability matrix P[s, a, s']-------------------------###

# Random generation of matrix
def Markov_decision_matrix(num_states, num_actions):
  Markov = []
  for s in range(num_states):
    matrix = np.random.rand(num_actions,num_states)
    # Normalization of each row
    matrix = matrix/matrix.sum(axis=1, keepdims=True)
    Markov.append(matrix)
  return np.array(Markov)

#------------Transition probabilities------------#

np.random.seed(42)
P = Markov_decision_matrix(num_states, num_actions)




##--------------------------Reward function R[s, a]--------------------------------------##

R = np.round(np.random.rand(num_states, num_actions),2)  # random generation of R

# Discount factor
gamma = 0.9

# Coefficient of v in objective function
mu = np.random.randint(1,num_actions, size = num_states)
```

```python
##-------------------Function to solve MDPs using linear programm------------##

def Linear_Programming(P,R,gamma, nb_states,nb_actions,disp = "F", mu = None):
  states = [f's{i}' for i in range(nb_states)]
  actions = [f'a{i}' for i in range(nb_actions)]

    # Initializing a linear programming problem
  lp = pulp.LpProblem("MDP", pulp.LpMinimize)

  # Definition of the value variables
  V = pulp.LpVariable.dicts("V", states, lowBound=None, cat='Continuous')

  # Objective function: minimize the sum of state values
  if mu is not None:
    lp += pulp.lpSum([mu[s] * V[states[s]] for s in range(nb_states)])
  else:
    lp += pulp.lpSum([V[s] for s in states]), "Objective"

  # Adding constraints for each state and action
  for s in range(nb_states):
      for a in range(nb_actions):
          lp += V[states[s]] >= R[s,a] + gamma * pulp.lpSum([P[s,a,s_prime] * V[states[s_prime]] for s_prime in range(nb_states)]) #, f"Constraint_{s}_{a}"

  # Solving the linear program
  cplex_path = "/path/to/CPLEX_StudioXXX/cplex/bin/x86-64_linux/cplex"

  solver1 = pulp.CPLEX_CMD(path=cplex_path)
  solver2 = pulp.PULP_CBC_CMD()
  #lp.solve(solver1)
  lp.solve(solver2)


  if disp=="T":

    # Print the optimal values
    for s in states:
        print(f"V({s})= {pulp.value(V[s])}")

    # Deriving the optimal policy
    optimal_policy = {}
    for s in range(nb_states):
        best_action = None
        best_value = float('-inf')
        for a in range(nb_actions):
            value = R[s, a] + gamma * sum(P[s,a,s_prime] * pulp.value(V[states[s_prime]]) for s_prime in range(nb_states))
            if value > best_value:
                best_value = value
                best_action = a
        optimal_policy[states[s]] = best_action

    print("\n Optimal policy:")
    for s in range(nb_states):
        print(f"π{s}: a{optimal_policy[states[s]]}")

  return lp,V
```

```python
##------------- Impact of gamma on the resolution of Linear Program ----------------##
Gamma = [0.2,0.5,0.7,1-1e-1, 1-1e-2, 1-1e-3,1-1e-4]
LP_times = []

for gam in Gamma:
  start = time.time()
  _ ,_ = Linear_Programming(P,R,gam, num_states,num_actions)
  end = time.time()
  LP_times.append(end-start)

Gamma_log = [np.log10(1/(1-g)) for g in Gamma] # Log(1/(1-gamma))
Times_log = [np.log10(t) for t in LP_times] # Log(times)
plt.figure()
for i in range(len(LP_times)):
  plt.text(Gamma_log[i],Times_log[i],f'{Gamma[i],round(LP_times[i],5)}')

plt.plot(Gamma_log,Times_log, "-*", label="$(\gamma,t=f(\gamma))$ \n Linear Program")
plt.xlabel("$\log[\dfrac{1}{1-\gamma}]$")
plt.ylabel("$\log(t)$")
plt.title("Times to compute $V^*$ when $\gamma$ increases")
plt.legend()
plt.grid(1)
```

```python
###----------------------Analyze of convergence by varying the number of state--------------------###

LPState_times = []
S = [10,20,30,50,100,150,200,300,400,500] #states
a = 10 # fixed action
gamma = 0.7


for s in S:
  Pr = Markov_decision_matrix(s, a)
  Rr = np.round(np.random.rand(s, a),2)
  start = time.time()
  _ ,_ = Linear_Programming(Pr,Rr,gamma, s,a)
  end = time.time()
  LPState_times.append(end-start)

plt.figure()

plt.plot(S,LPState_times, "-*", label="$Linear Programing$")
plt.xlabel("$S$")
plt.ylabel("$CPU_{time}(s)$")
plt.title("Times to compute $V^*$ when $S$ increases")
plt.legend()
plt.grid(1)
```

```python
plt.figure()
Sss = [s**3 for s in S]
plt.plot(Sss,LPState_times, "-*", label="")
plt.xlabel("$S^3$")
plt.ylabel("$CPU_{time}(s)$")
plt.grid(1)
plt.legend()
```
